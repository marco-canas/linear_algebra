## Las fuerzas que pueden actuar sobre un punto

Consideremos un punto material $P$ y las fuerzas que pueden actuar sobre l. 

Si $F_{1}$ y $F_{2}$ son dos fuerzas que actan sobre $P$, se puede encontrar una fuerza $F$ resultante de $F_{1}$ y $F_{2}$ con la propiedad de ejercer sobre $P$ la misma accin que $F_{1}$ y $F_{2}$  actuando simultneamente. 

As que tenemos una fuerza $F$, que resulta de sumar $F_{1}$ y $F_{2}$, ($F = F_{1} + F_{2}$).

Veamos qu propiedades tiene esta operacin.  


Representemos $F_{1}$ y $F_{2}$  por vectores cuya direccin y sentido corresponden a los de las fuerzas dadas y su longitud (en una escala determinada) mide la magnitud; entonces, la resultante queda determinada por la diagonal del paralelogramo, dos de cuyos lados son los segmentos correspondientes a $F_{1}$ y $F_{2}$
(ver figura siguiente).

````python
import numpy as np
import matplotlib.pyplot as plt

# Definicin de las fuerzas como vectores
F1 = np.array([4, 1])   # Fuerza F1
F2 = np.array([2, 3])   # Fuerza F2

# Resultante (diagonal del paralelogramo)
R = F1 + F2

# Origen
O = np.array([0, 0])

# Crear la figura
plt.figure(figsize=(6, 6))

# Dibujar los vectores F1 y F2 desde el origen
plt.quiver(*O, *F1, angles='xy', scale_units='xy', scale=1, color='blue', label='$F_1$')
plt.quiver(*O, *F2, angles='xy', scale_units='xy', scale=1, color='green', label='$F_2$')

# Dibujar los lados del paralelogramo
plt.plot([F1[0], R[0]], [F1[1], R[1]], 'k--', linewidth=1)
plt.plot([F2[0], R[0]], [F2[1], R[1]], 'k--', linewidth=1)
plt.scatter(*O, color='black', s = 100)  # Punto de origen

# Dibujar la resultante
plt.quiver(*O, *R, angles='xy', scale_units='xy', scale=1, color='red', label='Resultante')

# Etiquetas
plt.text(F1[0], F1[1], '$F_1$', fontsize=12)
plt.text(F2[0], F2[1], '$F_2$', fontsize=12)
plt.text(R[0], R[1], '$R$', fontsize=12)

# Configuracin del grfico
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True)
plt.axis('equal')
plt.xlim(0, 7)
plt.ylim(0, 6)
plt.legend()
plt.title('Suma de fuerzas: mtodo del paralelogramo')

plt.show()
````

Sea $T$ el conjunto de todas las fuerzas que actan sobre $P$; entonces, podemos decir que:  


1. Si $F_{1}$ y $F_{2}$ pertenecen a $T$, entonces $F_{1} + F_{2} = F$ pertenece a $T$. 

A esta propiedad la llamaremos Clausuratividad o Cerradura de la adicin en $T$:

2. Asociatividad de la adicin en $T$ 
   $$
   (F_{1} + F_{2}) + F_{3} = F_{1} + (F_{2} + F_{3}) 
   $$

````python
import numpy as np
import matplotlib.pyplot as plt

# Definicin de los vectores
F1 = np.array([3, 1])
F2 = np.array([1, 3])
F3 = np.array([2, -1])

# Sumas parciales
S12 = F1 + F2          # F1 + F2
S23 = F2 + F3          # F2 + F3

# Resultante total
F = F1 + F2 + F3

# Origen
O = np.array([0, 0])

plt.figure(figsize=(7, 7))

# --- Vectores base ---
plt.quiver(*O, *F1, angles='xy', scale_units='xy', scale=1,
           color='blue', label='$F_1$')

plt.quiver(*F1, *F2, angles='xy', scale_units='xy', scale=1,
           color='green', label='$F_2$')

plt.quiver(*(F1 + F2), *F3, angles='xy', scale_units='xy', scale=1,
           color='purple', label='$F_3$')

# --- Suma F1 + F2 desde el origen ---
plt.quiver(*O, *S12, angles='xy', scale_units='xy', scale=1,
           color='orange', linestyle='dashed', linewidth=2,
           label='$F_1 + F_2$')

# --- Suma F2 + F3 aplicada en el extremo de F1 ---
plt.quiver(*F1, *S23, angles='xy', scale_units='xy', scale=1,
           color='brown', linestyle='dashed', linewidth=2,
           label='$F_2 + F_3$ (desde $F_1$)')

# --- Resultante final ---
plt.quiver(*O, *F, angles='xy', scale_units='xy', scale=1,
           color='red', linewidth=3, label='Resultante $F$')

# Etiquetas
plt.text(F1[0], F1[1]-0.2, '$F_1$', fontsize=11)
plt.text((F1 + F2 / 2)[0]+0.2, (F1 + F2 / 2)[1], '$F_2$', fontsize=11)
plt.text(*(F1 + F2 + F3 / 2), '$F_3$', fontsize=11)
plt.text(*S12, '$F_1+F_2$', fontsize=11)
plt.text(*(F1 + S23), '$F_2+F_3$', fontsize=11)
plt.text(*F, '$F$', fontsize=12)

# Configuracin del grfico
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True)
plt.axis('equal')
plt.xlim(-1, 8)
plt.ylim(-1, 6)
plt.legend()
plt.title('Asociatividad de la suma de vectores\n$(F_1 + F_2) + F_3 = F_1 + (F_2 + F_3)$')

plt.savefig(r"C:\Users\marco\Downloads\asociatividad_suma_vectores.png")
plt.show()
````

3. Modulatividad de la adicin en $T$: Por otra parte, existe una fuerza de magnitud cero, fuerza nula, que podemos designar $0$ y tiene la propiedad de que:

$$
F+0 = 0 + F = F
$$

Es decir, la adicin en $T$ tiene un elemento neutro, la fuerza nula.

4. Ahora, si llamamos la opuesta de $F$ a una fuerza $F'$ que tiene la misma magnitud y direccin que $F$ pero sentido contrario, encontramos que la resultante de  $F$ y  $F'$ es cero, es decir:  

Para cada $F$ en $T$ existe ,$F'$ tal que $F + F' = 0$. Esta $F'$ se llamar $-F$.  

A esta propiedad de la adicin en $T$ la llamaremos existencia del inverso aditivo o invertivilidad de la adicin en $T$.  

````python
import numpy as np
import matplotlib.pyplot as plt

# Definicin de la fuerza
F = np.array([4, 2])     # Fuerza F
minus_F = -F             # Inverso aditivo de F

# Origen
O = np.array([0, 0])

plt.figure(figsize=(6, 6))

# Vector F
plt.quiver(*O, *F, angles='xy', scale_units='xy', scale=1,
           color='blue', linewidth=2, label='$F$')

# Vector -F
plt.quiver(*O, *minus_F, angles='xy', scale_units='xy', scale=1,
           color='red', linewidth=2, label='$-F$')

# Etiquetas
plt.text(*F, '$F$', fontsize=12)
plt.text(minus_F[0] , minus_F[1] - 0.2, '$-F$', fontsize=12)

# Configuracin del plano
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True)
plt.axis('equal')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.legend()
plt.title('Fuerza $F$ y su inverso aditivo $-F$')

plt.show()
````

5. Por ltimo, es claro que el orden en que se consideran las fuerzas no altera la
resultante, esto es:  

$$
F_{1} + F_{2} = F_{2} + F_{1}
$$

A esta propiedad de la adicin en $T$ la llamaremos conmutatividad de la adicin en $T$.

````python
import numpy as np
import matplotlib.pyplot as plt

# Definicin de las fuerzas
F1 = np.array([4, 1])
F2 = np.array([1, 3])

# Resultante comn
R = F1 + F2

# Origen
O = np.array([0, 0])

plt.figure(figsize=(7, 7))

# --- Recorrido 1: F1 + F2 ---
plt.quiver(*O, *F1, angles='xy', scale_units='xy', scale=1,
           color='blue', linewidth=2, label='$F_1$')

plt.quiver(*F1, *F2, angles='xy', scale_units='xy', scale=1,
           color='green', linewidth=2, label='$F_2$')

# --- Recorrido 2: F2 + F1 ---
plt.quiver(*O, *F2, angles='xy', scale_units='xy', scale=1,
           color='green', linestyle='dashed', linewidth=2)

plt.quiver(*F2, *F1, angles='xy', scale_units='xy', scale=1,
           color='blue', linestyle='dashed', linewidth=2)

# --- Resultante ---
plt.quiver(*O, *R, angles='xy', scale_units='xy', scale=1,
           color='red', linewidth=3, label='Resultante')

# Etiquetas
plt.text(*F1, '$F_1$', fontsize=11)
plt.text(*F2, '$F_2$', fontsize=11)
plt.text(*R, '$R$', fontsize=12)

# Configuracin del grfico
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True)
plt.axis('equal')
plt.xlim(-1, 7)
plt.ylim(-1, 6)
plt.legend()
plt.title('Conmutatividad de la suma de fuerzas\n$F_1 + F_2 = F_2 + F_1$')

plt.show()
````
