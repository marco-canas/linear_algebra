# Definicin 1  



Sean $v_{1}, v_{2}, \ldots, v_{n}$ vectores en un espacio vectorial $V$. Decimos que $x$ es una combinacin lineal de los vectores $v_{1}, v_{2}, \ldots, v_{n}$ si existen escalares $a_{1}, a_{2}, \ldots, a_{n}$, tales que:



$$
x = a_{1}v_{1} + a_{2}v_{2} + \ldots + a_{n}v_{n}
$$
es decir, si existen escalares $a_{1}, a_{2}, \ldots, a_{n}$ tales que:
$$
x = \sum_{i=1}^{n} a_{i}v_{i}$$ 
$$

# Observaciones  

1. Toda combinacin lineal de vectores de un espacio dado es otro vector del mismo espacio. Es decir, si combinas vectores de dos componentes, te da un vector de dos componente. 

2.  El mdulo de un espacio vectorial es combinacin lineal de cualquier conjunto de
    vectores del mismo espacio.  

$$
\mathbf{0} = 0v_{1} + 0v_{2} + \cdots + 0v_{3} 
$$

3. Todo vector es combinacin de s mismo.  

$$
x = 1x
$$

# Ejemplos  

1. La figura siguiente muestra un vector $v$ en $\mathbb{R}^{2}$ o $\mathbb{R}^{3}$ como combinacin lineal de los vectores $v_{1}$ y $v_{2}$.  

$$
v = c_{1} v_{1} + c_{2}v_{2} 
$$

```python
# dame con matplotlib el grafico de dos vectores v1 y v1, 
# y el grfico en el mismo plano de los mltiplos escalares c1v1 y c2v2,
# con escalares c1 y c2 positivos. Adems inclute el trazo del vector 
# v = c1v1+c2v2 y el paralelogramo determinado por 
# c1v1 y c2v2  

import numpy as np
import matplotlib.pyplot as plt

# Definicin de los vectores
v1 = np.array([2, 1])
v2 = np.array([1, 3])

# Escalares positivos
c1 = 2
c2 = 1.5

# Mltiplos escalares
c1v1 = c1 * v1
c2v2 = c2 * v2

# Vector resultante
v = c1v1 + c2v2

# Crear la figura
plt.figure(figsize=(8, 8))
plt.axhline(0)
plt.axvline(0)
plt.grid(True)

# Vectores originales
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green')

# Mltiplos escalares
plt.quiver(0, 0, c1v1[0], c1v1[1], angles='xy', scale_units='xy', scale=1, color='cyan')
plt.quiver(0, 0, c2v2[0], c2v2[1], angles='xy', scale_units='xy', scale=1, color='lime')

# Vector suma
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red')

# Paralelogramo
x_par = [0, c1v1[0], v[0], c2v2[0], 0]
y_par = [0, c1v1[1], v[1], c2v2[1], 0]
plt.plot(x_par, y_par, 'k--')

# Etiquetas
plt.text(v1[0], v1[1], r'$v_1$', color='blue')
plt.text(v2[0], v2[1], r'$v_2$', color='green')
plt.text(c1v1[0], c1v1[1], r'$c_1v_1$', color='cyan')
plt.text(c2v2[0], c2v2[1], r'$c_2v_2$', color='lime')
plt.text(v[0], v[1], r'$v=c_1v_1+c_2v_2$', color='red')

# Ajustes finales
plt.axis('equal')
plt.xlim(0, 7)
plt.ylim(0, 7)
plt.title('Combinacin lineal y paralelogramo generado')
plt.savefig(r"C:\Users\marco\Downloads\Combinacin_lineal_y_paralelogramo_generado.png")
plt.show()
```

2.  Si  $a$ y $b$ son vectores libres, $Pr(a/b)$, la proyeccin de $a$ sobre $b$ es una combinacin lineal de $b$ y $Pr(b/a)$ es una combinacin lineal de $a$.

```python
# traza dos vectores a y b y a las proyecciones de a sobre b y la proyeccin de b sobre a en un 
# mismo plano con matplotlib.  

import numpy as np
import matplotlib.pyplot as plt

# Definicin de los vectores
a = np.array([4, 1])
b = np.array([2, 3])

# Producto punto
dot_ab = np.dot(a, b)

# Proyecciones
proj_a_on_b = (dot_ab / np.dot(b, b)) * b
proj_b_on_a = (dot_ab / np.dot(a, a)) * a

# Crear figura
plt.figure(figsize=(8, 8))
plt.axhline(0)
plt.axvline(0)
plt.grid(True)

# Vectores originales
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1)

# Proyecciones
plt.quiver(0, 0, proj_a_on_b[0], proj_a_on_b[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, proj_b_on_a[0], proj_b_on_a[1], angles='xy', scale_units='xy', scale=1)

# Segmentos auxiliares (perpendiculares)
plt.plot([a[0], proj_a_on_b[0]], [a[1], proj_a_on_b[1]], '--')
plt.plot([b[0], proj_b_on_a[0]], [b[1], proj_b_on_a[1]], '--')

# Etiquetas
plt.text(a[0], a[1], r'$a$', fontsize=12)
plt.text(b[0], b[1], r'$b$', fontsize=12)
plt.text(proj_a_on_b[0], proj_a_on_b[1], r'$\mathrm{proj}_b(a)$', fontsize=12)
plt.text(proj_b_on_a[0], proj_b_on_a[1], r'$\mathrm{proj}_a(b)$', fontsize=12)

# Ajustes visuales
plt.axis('equal')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.title('Proyeccin de vectores en el plano')
plt.savefig(r"C:\Users\marco\Downloads\Proyeccin_de_vectores_en_el_plano.png")
plt.show()
```
