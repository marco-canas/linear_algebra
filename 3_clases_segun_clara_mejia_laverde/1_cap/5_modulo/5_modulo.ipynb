```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Dominio
x = np.linspace(-5, 5, 400)

# Funcin general
def f(x, n):
    return x**n

# Funcin de actualizacin
def plot_function(n):
    plt.figure(figsize=(7, 5))
    
    if n == 0:
        y = np.ones_like(x)
        label = r"$f(x)=1$"
    else:
        y = f(x, n)
        label = rf"$f(x)=x^{n}$"
    
    plt.plot(x, y, linewidth=2)
    plt.axhline(0)
    plt.axvline(0)
    plt.grid(True)
    plt.legend([label])
    plt.title(f"Representacin de la funcin para n = {n}")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.show()

# Deslizador
slider_n = widgets.IntSlider(
    value=1,
    min=0,
    max=3,
    step=1,
    description='n',
    continuous_update=False
)

# Interactividad
widgets.interact(plot_function, n=slider_n)
```

# 5.1 Definicin y ejemplos de bases   



# Definicin 1  

Un conjunto finito de vectores $\{v_{1}, v_{2}, \ldots, v_{n}\}$ es una *base* para un espacio vectorial $V$ si:  

1. $\{v_{1}, v_{2}, \ldots, v_{n}\}$ es linealmente independiente. 
2. $\{v_{1}, v_{2}, \ldots, v_{n}\}$ genera a $V$

# Ejemplos

1. Los vectores $e_{1} = (1,0)$ y $e_{1} = (0,1)$ forman una base para $\mathbb{R}^{2}$.  

```python
import matplotlib.pyplot as plt
import numpy as np

# Vectores de la base cannica
e1 = np.array([1, 0])
e2 = np.array([0, 1])

# Figura
plt.figure(figsize=(6, 6))

# Dibujo de vectores
plt.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1)

# Etiquetas
plt.text(1.05, 0, r"$e_1=(1,0)$", fontsize=12)
plt.text(0, 1.05, r"$e_2=(0,1)$", fontsize=12)

# Ejes
plt.axhline(0)
plt.axvline(0)
plt.grid(True)

# Configuracin del plano
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.gca().set_aspect('equal', adjustable='box')

# Ttulo
plt.title(r"Base cannica de $\mathbb{R}^2$")

plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

La dimensin de un espacio vectorial no nulo es el nmero de vectores en una base para $V$. Como el conjunto $\{0\}$ es linealmente dependiente, se dice que el espacio
vectorial $\{0\}$ tiene dimensin cero.

Cuando la base del espacio vectorial $V$ es un conjunto finito de vectores, se dice
que $V$ es de dimensin finita.

La dimensin de $V$ se denota por $dim\ V$.

# 5.4  El problema de la base



Este problema puede presentarse en una de estas dos formas:  

1. Construir una base para $V$, tomando los vectores de V. Si puede escoger un
conjunto de vectores que genere a $V$, puede eliminar los vectores LD, si los
hay, y obtener una base para $V$.  

2. Dado un conjunto $S$ de vectores de $V$, construir una base para $V$ aadiendo,
o bien eliminando, algunos vectores de $S$, o ambas cosas.  

Si $S$ genera a $V$, se procede como en el caso anterior. Si no, se aumenta $S$ hasta obtener un generador y luego se procede como en el caso anterior.

# Lnea argumentativa (Marco terico)



El desarrollo del **pensamiento computacional** constituye hoy un eje fundamental en la educacin matemtica. 

Como lo plantea **Wing (2006)**, esta forma de razonamiento debe considerarse una competencia transversal y bsica, al nivel de la lectura, la escritura y las matemticas. 

En consecuencia, fomentar su aprendizaje desde los niveles iniciales hasta la educacin superior resulta indispensable para formar profesionales capaces de enfrentar los retos de la sociedad contempornea.



En el campo de la didctica de las matemticas, **Papert (1980, 1996)** haba anticipado esta necesidad al concebir las computadoras no como simples herramientas de clculo, sino como **objetos para pensar**, capaces de favorecer la construccin activa de significados matemticos a travs de la programacin. 

Esta visin se alinea con el constructivismo de **Piaget**, que resalta el valor de la experimentacin en el aprendizaje, y con la perspectiva socio-constructivista de **Vygotsky**, que otorga un papel central a las herramientas culturales en la mediacin del conocimiento.



En este marco, el uso de entornos como **Jupyter Notebooks** y lenguajes como **Python, LaTeX, Markdown e ingls** no debe interpretarse como una rivalidad con la enseanza tradicional de la matemtica, sino como una **ampliacin de los registros de representacin** que describi **Duval (1993, 2006)**. 

Segn su teora semitica, el aprendizaje significativo se potencia cuando el estudiante logra traducir y articular distintos sistemas de representacin: grfico, algebraico, numrico, verbal y, en nuestro tiempo, digital.



De igual manera, el **Enfoque Ontosemitico de Godino (2002, 2017)** muestra que los significados matemticos emergen de prcticas diversas, lo que justifica la incorporacin de estos lenguajes como recursos que enriquecen las configuraciones de prctica matemtica, en sintona con las exigencias de la investigacin y la prctica profesional.



En el mbito internacional, organizaciones como el **NCTM (2000, 2014)** han subrayado que la tecnologa es esencial para la enseanza y el aprendizaje de la matemtica, dado que facilita la experimentacin, la validacin de conjeturas y la comunicacin de ideas. 

Asimismo, instancias como la **ACM** y el **IEEE** han sealado la importancia de desarrollar habilidades computacionales como requisito fundamental en la formacin universitaria, tanto en el sector pblico como privado.



Por tanto, ensear matemticas integrando Python, LaTeX, Markdown e ingls en ambientes como Jupyter Notebooks contribuye a la formacin de profesionales con **competencias cientficas, comunicativas y tecnolgicas**, capaces de comprender la matemtica de manera interactiva y experimental, y de responder a las demandas del mundo laboral. 

Esta integracin, lejos de ser inadecuada, constituye un **puente entre la enseanza acadmica y las necesidades de la sociedad**, fortaleciendo la pertinencia de los programas universitarios.
