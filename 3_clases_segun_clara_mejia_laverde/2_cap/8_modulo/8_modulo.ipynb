# 8.1 Conjuntos ortonormales  



## Definicin 1 (Definicin de subconjunto ortonormal)  

En $\mathbb{R}^{n}$ (con el producto interno usual), un **subconjunto ortonormal** es un conjunto de vectores que son mutuamente ortogonales y adems tienen norma unitaria.

Formalmente, un conjunto de vectores
$$
{v_1, v_2, \dots, v_k} \subset \mathbb{R}^n
$$
se dice **ortonormal** si cumple:

1. **Ortogonalidad:**
   $$
   v_i \cdot v_j = 0 \quad \text{si } i \neq j
   $$

2. **Normalizacin (norma 1):**
   $$
   |v_i| = 1 \quad \text{para todo } i
   $$

De manera equivalente, ambas condiciones se resumen como:
$$
v_i \cdot v_j =
\begin{cases}
1 & \text{si } i = j,\\  
0 & \text{si } i \neq j.
\end{cases}
$$

Es decir, los vectores del subconjunto son perpendiculares entre s y cada uno tiene longitud igual a 1.

Un ejemplo tpico en $\mathbb{R}^n$ es la base cannica:
$$
e_1=(1,0,\dots,0),\quad e_2=(0,1,\dots,0),\ \dots,\ e_n=(0,0,\dots,1),
$$
que forma un subconjunto ortonormal de $\mathbb{R}^n$.


# La norma en $\mathbb{R}^{n}$.  

En $\mathbb{R}^n$, la **norma** (o longitud) de un vector se determina usando la norma euclidiana, definida a partir del producto interno usual.

Si
$$
v=(v_1,v_2,\dots,v_n)\in \mathbb{R}^n,
$$
entonces su norma es:

$$
|v|=\sqrt{v_1^2+v_2^2+\cdots+v_n^2}.
$$

Equivalente mente, usando producto punto:

$$
|v|=\sqrt{v\cdot v}.
$$

### Interpretacin

La norma mide la longitud del vector en el espacio. Es la generalizacin del teorema de Pitgoras a $n$ dimensiones.

### Ejemplo

Si $v=(3,-1,2)\in \mathbb{R}^3$, entonces:

$$
|v|=\sqrt{3^2+(-1)^2+2^2}
=\sqrt{9+1+4}
=\sqrt{14}.
$$


Nota:   

los conceptos de ortogonalidad y magnitud de los vectores de n estn
dados en trminos del producto escalar. Es importante, antes de seguir adelante,
recordar las propiedades del producto escalar.

# Recordemos las propiedades del producto escalar en $\mathbb{R}^{n}$ 

En $\mathbb{R}^n$, el **producto escalar** (o producto punto) entre dos vectores
$v=(v_1,\dots,v_n)$ y $w=(w_1,\dots,w_n)$ se define como

$$
v\cdot w = v_1w_1 + v_2w_2 + \cdots + v_nw_n.
$$

Este producto satisface las siguientes propiedades fundamentales:

### 1. Conmutatividad

$$
v\cdot w = w\cdot v.
$$

### 2. Linealidad en cada argumento

Para $u,v,w\in\mathbb{R}^n$ y $\lambda\in\mathbb{R}$:

* Respecto a la suma:
  $$
  (v+u)\cdot w = v\cdot w + u\cdot w,
  $$
  $$
  v\cdot (w+u) = v\cdot w + v\cdot u.
  $$

* Respecto al producto por escalar:
  $$
  (\lambda v)\cdot w = \lambda (v\cdot w),
  $$
  $$
  v\cdot (\lambda w) = \lambda (v\cdot w).
  $$

### 3. Positividad

$$
v\cdot v \ge 0 \quad \text{para todo } v\in\mathbb{R}^n.
$$

### 4. Definido positivo

$$
v\cdot v = 0 \iff v=0.
$$

### 5. Relacin con la norma

El producto escalar determina la norma euclidiana:
$$
|v|=\sqrt{v\cdot v}.
$$

### 6. Ortogonalidad

Dos vectores son ortogonales si y solo si:
$$
v\cdot w = 0.
$$

Estas propiedades hacen del producto escalar la herramienta central para medir longitudes, ngulos y proyecciones en $\mathbb{R}^n$.


# Ejemplo 1 (Ejemplo de un conjunto ortogonal en $\mathbb{R}^{3}$ y el procedimiento para normalizarlo)  

Consideremos el siguiente conjunto de tres vectores en $\mathbb{R}^3$:

$$
v_1=(1,1,0), \quad v_2=(1,-1,0), \quad v_3=(0,0,2).
$$

### 1. Verificacin de que es un conjunto ortogonal

Calculamos los productos escalares:

$$
v_1\cdot v_2 = 1\cdot 1 + 1\cdot(-1) + 0\cdot 0 = 1-1+0=0,
$$

$$
v_1\cdot v_3 = 1\cdot 0 + 1\cdot 0 + 0\cdot 2 = 0,
$$

$$
v_2\cdot v_3 = 1\cdot 0 + (-1)\cdot 0 + 0\cdot 2 = 0.
$$

Como todos los productos cruzados son cero, el conjunto es **ortogonal**.

---

### 2. Clculo de las normas

$$
|v_1|=\sqrt{1^2+1^2+0^2}=\sqrt{2},
$$

$$
|v_2|=\sqrt{1^2+(-1)^2+0^2}=\sqrt{2},
$$

$$
|v_3|=\sqrt{0^2+0^2+2^2}=2.
$$

---

### 3. Procedimiento de normalizacin

Para normalizar un vector se divide por su norma:

$$
u_i=\frac{v_i}{|v_i|}.
$$

Aplicando esto:

$$
u_1=\frac{1}{\sqrt{2}}(1,1,0)=\left(\frac{1}{\sqrt{2}},\frac{1}{\sqrt{2}},0\right),
$$

$$
u_2=\frac{1}{\sqrt{2}}(1,-1,0)=\left(\frac{1}{\sqrt{2}},-\frac{1}{\sqrt{2}},0\right),
$$

$$
u_3=\frac{1}{2}(0,0,2)=(0,0,1).
$$

---



### 4. Conjunto ortonormal obtenido

El conjunto normalizado es:

$$
\left\{
\left(\frac{1}{\sqrt{2}},\frac{1}{\sqrt{2}},0\right),
\left(\frac{1}{\sqrt{2}},-\frac{1}{\sqrt{2}},0\right),
(0,0,1)
\right\},
$$

el cual es un **conjunto ortonormal** en $\mathbb{R}^3$ porque:

* los vectores siguen siendo ortogonales entre s,
* y cada uno tiene norma igual a $1$.

# 8.2 Otra forma del problema de la base. Proceso de ortonormalizacin de Gram - Schmidt   

Planteamiento del problema

Dada una base { } 1 2 , ,..., S= v v vk  para un subespacio de ,n hallar una base
ortonormal { } 1 2 ,..., . k B=u,u u La solucin a este problema nos la brinda el siguiente teorema, conocido como proceso de ortonormalizacin de Gram-Schmidt



El **proceso de ortonormalizacin de GramSchmidt** permite transformar una base cualquiera de un subespacio de $\mathbb{R}^n$ en una **base ortonormal** del mismo subespacio, conservando el espacio generado.

---



## Planteamiento general

Sea
$$
S=\{v_1,v_2,\dots,v_k\}
$$
una base de un subespacio $W\subseteq\mathbb{R}^n$ (vectores linealmente independientes).

El objetivo es construir un conjunto
$$
B=\{u_1,u_2,\dots,u_k\}
$$
tal que:

* genere el mismo subespacio $W$,
* los vectores sean ortogonales entre s,
* y cada vector tenga norma 1 (ortonormalidad).

El procedimiento tiene dos etapas: **ortogonalizacin** y **normalizacin**.

---



# Paso 1: Construccin de un sistema ortogonal

Se construyen vectores $w_1,w_2,\dots,w_k$ ortogonales entre s a partir de los $v_i$.

### Primer vector

Se toma directamente:
$$
w_1 = v_1.
$$



### Segundo vector

Se elimina de $v_2$ la componente en la direccin de $w_1$:

$$
w_2 = v_2 - \operatorname{proj}_{w_1}(v_2),
$$

donde la proyeccin es:

$$
\operatorname{proj}_{w_1}(v_2)=
\frac{v_2\cdot w_1}{w_1\cdot w_1}w_1.
$$

---


````python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# Vectores en el plano
# =========================
v1 = np.array([4, 2])     # ms largo
v2 = np.array([1, 3])

# =========================
# Proyeccin de v2 sobre v1
# =========================
proj_v2_v1 = (np.dot(v2, v1) / np.dot(v1, v1)) * v1

# Vector ortogonal de Gram-Schmidt
w2 = v2 - proj_v2_v1

# =========================
# Crear figura
# =========================
plt.figure(figsize=(7,7))

# Ejes coordenados
plt.axhline(0)
plt.axvline(0)

# =========================
# Dibujar vectores
# =========================
plt.quiver(0, 0, v1[0], v1[1],
           angles='xy', scale_units='xy', scale=1,
           color='blue', linewidth=3, label=r'$v_1$')

plt.quiver(0, 0, v2[0], v2[1],
           angles='xy', scale_units='xy', scale=1,
           color='green', linewidth=3, label=r'$v_2$')

plt.quiver(0, 0, proj_v2_v1[0], proj_v2_v1[1],
           angles='xy', scale_units='xy', scale=1,
           color='red', linewidth=3,
           label=r'$\mathrm{proy}_{v_1}(v_2)$')

# Vector ortogonal desde la proyeccin hasta la punta de v2
plt.quiver(proj_v2_v1[0], proj_v2_v1[1], w2[0], w2[1],
           angles='xy', scale_units='xy', scale=1,
           color='purple', linewidth=3,
           label=r'$w_2 = v_2 - \mathrm{proy}_{v_1}(v_2)$')

# =========================
# Trazo de la proyeccin
# =========================
plt.plot([v2[0], proj_v2_v1[0]],
         [v2[1], proj_v2_v1[1]],
         linestyle='dashed',
         color='black',
         linewidth=2,
         label='Trazo de la proyeccin')

# =========================
# Ajustes visuales
# =========================
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.gca().set_aspect('equal', adjustable='box')

plt.title('Paso de GramSchmidt en el plano')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.grid(True)

plt.show()
````


### Tercer vector

Se eliminan las componentes en las direcciones de $w_1$ y $w_2$:

$$
w_3 = v_3

* \frac{v_3\cdot w_1}{w_1\cdot w_1}w_1
* \frac{v_3\cdot w_2}{w_2\cdot w_2}w_2.
  $$

---

### Caso general

Para $j=1,\dots,k$:

$$
w_j
===

## v_j

\sum_{i=1}^{j-1}
\frac{v_j\cdot w_i}{w_i\cdot w_i},w_i.
$$

Con esto se obtiene un conjunto ortogonal:

$$
{w_1,\dots,w_k}.
$$

---

# Paso 2: Normalizacin

Se convierten los vectores ortogonales en unitarios:

$$
u_j=\frac{w_j}{|w_j|}, \quad j=1,\dots,k.
$$

Entonces:

$$
B={u_1,\dots,u_k}
$$

es una base ortonormal del subespacio $W$.

---

# Idea geomtrica del mtodo

En cada paso:

* Se toma un vector nuevo $v_j$.
* Se le quitan las partes que apuntan en las direcciones ya construidas.
* El resultado $w_j$ queda perpendicular a todos los anteriores.
* Finalmente se normaliza.

Es como ir construyendo ejes perpendiculares adaptados al subespacio generado por los vectores originales.

---

# Propiedades importantes

1. El subespacio generado no cambia:
   $$
   \text{span}(S)=\text{span}(B).
   $$

2. Los vectores obtenidos son ortogonales:
   $$
   w_i\cdot w_j=0 \quad (i\neq j).
   $$

3. La base final es ortonormal:
   $$
   u_i\cdot u_j=
   \begin{cases}
   1 & i=j,\
   0 & i\neq j.
   \end{cases}
   $$

4. El proceso funciona para cualquier base linealmente independiente en $\mathbb{R}^n$.

---



# Esquema operativo (receta prctica)

Dada una base $S={v_1,\dots,v_k}$:

1. $w_1=v_1$
2. Para $j=2,\dots,k$:
   $$
   w_j=v_j-\sum_{i=1}^{j-1}\frac{v_j\cdot w_i}{w_i\cdot w_i}w_i
   $$
3. Normalizar:
   $$
   u_j=\frac{w_j}{|w_j|}
   $$
4. La base ortonormal buscada es:
   $$
   B={u_1,\dots,u_k}.
   $$

Este es el procedimiento estndar de GramSchmidt para construir bases ortonormales en $\mathbb{R}^n$.


# 8.3 Proyecciones ortogonales 

## Definicin 2 (Proyeccin ortogonal de un vector v sobre un subespacio H de Rn)  

Sea $H$ un subespacio de $\mathbb{R}^n$ y sea $v\in\mathbb{R}^n$.
La **proyeccin ortogonal de $v$ sobre $H$** es el vector de $H$ que est ms cerca de $v$ y cuya diferencia con $v$ es perpendicular a todo el subespacio.



Formalmente, el vector $\operatorname{proj}_H(v)\in H$ es la proyeccin ortogonal de $v$ sobre $H$ si satisface:

1. $\operatorname{proj}_H(v)\in H$,
2. $v-\operatorname{proj}_H(v)\perp H$, es decir,
   $$
   (v-\operatorname{proj}_H(v))\cdot h=0 \quad \text{para todo } h\in H.
   $$



Equivalente mente, si llamamos $p=\operatorname{proj}_H(v)$, entonces existe una descomposicin nica:

$$
v=p+w,
$$

donde

$$
p\in H \quad \text{y} \quad w\in H^\perp.
$$

---



### Expresin mediante una base ortonormal

Si ${u_1,u_2,\dots,u_k}$ es una base ortonormal de $H$, entonces la proyeccin ortogonal de $v$ sobre $H$ est dada por:

$$
\operatorname{proj}_H(v)
=
(v\cdot u_1)u_1
+
(v\cdot u_2)u_2
+\cdots+
(v\cdot u_k)u_k.
$$

Esta frmula muestra que la proyeccin se obtiene sumando las componentes de $v$ en las direcciones ortonormales que generan el subespacio $H$.


Veamos un ejemplo concreto en $\mathbb{R}^3$ donde el subespacio $H$ es un plano.

---

# Ejemplo en $\mathbb{R}^3$

Consideremos el plano (subespacio)

$$
H={(x,y,z)\in\mathbb{R}^3:; x+y+z=0},
$$

que pasa por el origen y tiene como vector normal

$$
n=(1,1,1).
$$

Tomemos el vector

$$
v=(2,1,3).
$$


# Primera forma de solucin de este problema  

A continuacin presentamos **otra forma de obtener** la proyeccin ortogonal de
$v=(2,1,3)$ sobre el plano

$$
H={(x,y,z)\in\mathbb{R}^3:;x+y+z=0},
$$

siguiendo tres pasos:

1. hallar una base del plano,
2. ortonormalizarla (GramSchmidt),
3. calcular la proyeccin usando la base ortonormal.

---



# 1. Hallar una base del plano (H)

El plano est dado por la ecuacin:

$$
x+y+z=0.
$$

Despejamos, por ejemplo, $x=-y-z$:

$$
(x,y,z)=(-y-z,y,z)=y(-1,1,0)+z(-1,0,1).
$$

Por tanto, una base de (H) es:

$$
\mathcal{B}={w_1=(-1,1,0),; w_2=(-1,0,1)}.
$$

---



# 2. Ortonormalizar la base (GramSchmidt)

## Paso 1

Tomamos

$$
u_1=\frac{w_1}{|w_1|}.
$$

Como

$$
|w_1|=\sqrt{(-1)^2+1^2+0^2}=\sqrt{2},
$$

entonces

$$
u_1=\frac{1}{\sqrt{2}}(-1,1,0).
$$

---



## Paso 2

Ortogonalizamos $w_2$ respecto a $u_1$:

$$
\operatorname{proj}_{u_1}(w_2)=(w_2\cdot u_1)u_1.
$$

Calculamos:

$$
w_2\cdot u_1=
(-1,0,1)\cdot\frac{1}{\sqrt{2}}(-1,1,0)
=\frac{1}{\sqrt{2}}.
$$

Entonces

$$
\operatorname{proj}_{u_1}(w_2)
=\frac{1}{\sqrt{2}},u_1
=\frac{1}{2}(-1,1,0).
$$

Ahora:

$$
z_2=w_2-\operatorname{proj}_{u_1}(w_2)
=(-1,0,1)-\left(-\tfrac12,\tfrac12,0\right)
=\left(-\tfrac12,-\tfrac12,1\right).
$$

Su norma es:

$$
|z_2|=\sqrt{\tfrac14+\tfrac14+1}=\sqrt{\tfrac32}.
$$

Normalizamos:

$$
u_2=\frac{z_2}{|z_2|}
=\frac{1}{\sqrt{3/2}}\left(-\tfrac12,-\tfrac12,1\right)
=\left(-\frac{1}{\sqrt6},-\frac{1}{\sqrt6},\frac{2}{\sqrt6}\right).
$$

---



## Base ortonormal del plano

$$
{u_1,u_2}
$$

con

$$
u_1=\frac{1}{\sqrt2}(-1,1,0),\qquad
u_2=\left(-\frac{1}{\sqrt6},-\frac{1}{\sqrt6},\frac{2}{\sqrt6}\right).
$$

---



# 3. Proyeccin de (v) sobre (H)

Si ${u_1,u_2}$ es una base ortonormal de (H), entonces

$$
\operatorname{proj}_H(v)
=(v\cdot u_1)u_1+(v\cdot u_2)u_2.
$$

---



## Clculo de productos escalares

### Con (u_1)

$$
v\cdot u_1
=(2,1,3)\cdot\frac{1}{\sqrt2}(-1,1,0)
=\frac{-2+1}{\sqrt2}
=-\frac{1}{\sqrt2}.
$$

Entonces

$$
(v\cdot u_1)u_1
=-\frac{1}{\sqrt2}\cdot\frac{1}{\sqrt2}(-1,1,0)
=\left(\tfrac12,-\tfrac12,0\right).
$$

---



### Con (u_2)

$$
v\cdot u_2
=(2,1,3)\cdot\left(-\frac{1}{\sqrt6},-\frac{1}{\sqrt6},\frac{2}{\sqrt6}\right)
=\frac{-2-1+6}{\sqrt6}
=\frac{3}{\sqrt6}.
$$

Por tanto,

$$
(v\cdot u_2)u_2
=\frac{3}{\sqrt6}
\left(-\frac{1}{\sqrt6},-\frac{1}{\sqrt6},\frac{2}{\sqrt6}\right)
=\left(-\tfrac12,-\tfrac12,1\right).
$$

---



## Suma final

$$
\operatorname{proj}_H(v)
=\left(\tfrac12,-\tfrac12,0\right)
+\left(-\tfrac12,-\tfrac12,1\right)
=(0,-1,1).
$$

---



# Resultado

$$
\boxed{\operatorname{proj}_H(v)=(0,-1,1)}
$$

que coincide con el obtenido usando el vector normal.

Adems,

$$
0+(-1)+1=0,
$$

por lo que efectivamente pertenece al plano (H).



## Proyeccin ortogonal de (v) sobre el plano (H)

La proyeccin sobre un plano que pasa por el origen puede calcularse restando la componente de (v) en la direccin del vector normal:

$$
\operatorname{proj}_H(v)=v-\operatorname{proj}_n(v),
$$

donde

$$
\operatorname{proj}_n(v)=\frac{v\cdot n}{n\cdot n},n.
$$

Calculamos:

$$
v\cdot n = 2+1+3=6,\qquad n\cdot n=1+1+1=3.
$$

Entonces

$$
\operatorname{proj}_n(v)=\frac{6}{3}(1,1,1)=(2,2,2).
$$

Finalmente,

$$
\operatorname{proj}_H(v)=v-(2,2,2)=(2,1,3)-(2,2,2)=(0,-1,1).
$$

Y se verifica que pertenece al plano:

$$
0+(-1)+1=0.
$$

---

# Teorema 3  (Expresin de una vector $v$ como suma de sus proyecciones con respecto a los vectores de una base ortonormal $B$ para $\mathbb{R}^{n}$)  

**Teorema (descomposicin respecto de una base ortonormal de (\mathbb{R}^n)).**

Sea
$$
B={u_1,u_2,\dots,u_n}
$$
una base ortonormal de $\mathbb{R}^n$, y sea $v\in\mathbb{R}^n$.
Entonces $v$ puede expresarse de manera nica como suma de sus proyecciones sobre los vectores de la base:

$$
v=\sum_{i=1}^{n}(v\cdot u_i),u_i.
$$

Adems, cada trmino

$$
(v\cdot u_i),u_i
$$

es la proyeccin ortogonal de $v$ sobre la recta generada por $u_i$:

$$
\operatorname{proj}_{\langle u_i\rangle}(v)=(v\cdot u_i),u_i.
$$

En consecuencia,

$$
v=\operatorname{proj}*{\langle u_1\rangle}(v)
+\operatorname{proj}*{\langle u_2\rangle}(v)
+\cdots+
\operatorname{proj}_{\langle u_n\rangle}(v).
$$

Esta descomposicin es nica y los coeficientes

$$
v\cdot u_1,; v\cdot u_2,; \dots,; v\cdot u_n
$$

son precisamente las coordenadas de $v$ respecto de la base ortonormal $B$.


# Ejemplo 3  

Veamos un ejemplo concreto del teorema en (\mathbb{R}^3) y su visualizacin con Plotly.

---

# Ejemplo en (\mathbb{R}^3)

Tomemos la base ortonormal cannica:

$$
B={e_1,e_2,e_3},\qquad
e_1=(1,0,0),; e_2=(0,1,0),; e_3=(0,0,1).
$$

Sea el vector

$$
v=(2,1,3)\in\mathbb{R}^3.
$$

## Proyecciones de (v) sobre cada vector de la base

Como la base es ortonormal:

$$
\operatorname{proj}_{\langle e_i\rangle}(v)=(v\cdot e_i)e_i.
$$

Calculamos:

[
v\cdot e_1=2,\quad
v\cdot e_2=1,\quad
v\cdot e_3=3.
]

Por tanto,

[
\operatorname{proj}*{\langle e_1\rangle}(v)=2e_1=(2,0,0),
]
[
\operatorname{proj}*{\langle e_2\rangle}(v)=1e_2=(0,1,0),
]
[
\operatorname{proj}_{\langle e_3\rangle}(v)=3e_3=(0,0,3).
]

## Descomposicin del vector

[
v=(2,0,0)+(0,1,0)+(0,0,3).
]

Es decir,

[
v=\sum_{i=1}^3 (v\cdot e_i)e_i,
]

que es exactamente el enunciado del teorema.


# 8.4 Complemento ortogonal  



# Definicin $3$  

Sea $H$ un subespacio de $\mathbb{R}^n$ con el producto escalar usual.

El **complemento ortogonal** de $H$, denotado por $H^\perp$, se define como el conjunto de todos los vectores de $\mathbb{R}^n$ que son ortogonales a cada vector de $H$:

$$
H^\perp=\{v\in\mathbb{R}^n:\ v\cdot h=0 \text{ para todo } h\in H\}.
$$

Es decir, un vector pertenece a $H^\perp$ si su producto escalar con cualquier vector del subespacio $H$ es cero.

---



### Propiedades fundamentales



1. $H^\perp$ es un subespacio de $\mathbb{R}^n$.

2. Se cumple:
   $$
   H\cap H^\perp={0}.
   $$

3. Todo vector $v\in\mathbb{R}^n$ puede descomponerse de manera nica como
   $$
   v=p+q,
   $$

donde
   $$
   p\in H \quad \text{y} \quad q\in H^\perp.
   $$



4. Las dimensiones satisfacen:
   $$
   \dim(H)+\dim(H^\perp)=n.
   $$

5. Se cumple:
   $$
   (H^\perp)^\perp=H.
   $$



Geomtricamente, $H^\perp$ contiene todas las direcciones perpendiculares al subespacio $H$; por ejemplo, si $H$ es un plano en $\mathbb{R}^3$, entonces $H^\perp$ es la recta generada por el vector normal a ese plano.



## Ejemplo en $\mathbb{R}^2$



Sea el subespacio

$$
H = \text{la recta generada por } h = \begin{pmatrix} 2 \\  1 \end{pmatrix}.
$$

* Entonces $H = {\alpha h : \alpha \in \mathbb{R}}$.


* El **complemento ortogonal** $H^\perp$ es la recta formada por todos los vectores ortogonales a $h$, es decir, todos los vectores $v$ que satisfacen

$$
v \cdot h = 0.
$$

* Un vector que genera $H^\perp$ es $w = \begin{pmatrix}-1 \ 2 \end{pmatrix}$, porque

$$
h \cdot w = 2(-1) + 1(2) = -2 + 2 = 0.
$$

* Entonces

$$
H^\perp = {\beta w : \beta \in \mathbb{R}}.
$$

Geomtricamente, $H$ y $H^\perp$ son **rectas perpendiculares** que pasan por el origen.

---



## Visualizacin con matplotlib

Vamos a graficar:

* La recta $H$ en azul,
* Su complemento ortogonal $H^\perp$ en rojo,
* Un vector $v$ cualquiera descompuesto en $h + w$, con $h \in H$ y $w \in H^\perp$.

# Ejemplo en el espacio del concepto de complemento ortogoonal  



## Ejemplo en $\mathbb{R}^{3}$

Sea el subespacio $H \subset \mathbb{R}^{3}$ definido como el **plano que pasa por el origen** generado por los vectores:

$$
h_1 = \begin{pmatrix}1 \\  1 \\  0\end{pmatrix}, \quad
h_2 = \begin{pmatrix}0 \\  1 \\  1\end{pmatrix}.
$$



Entonces:

$$
H = \text{gen}{h_1, h_2}.
$$



El **complemento ortogonal** $H^\perp$ es el conjunto de todos los vectores ortogonales a cada vector de $H$:

$$
H^\perp = {v \in \mathbb{R}^3 : v \cdot h_1 = 0 \text{ y } v \cdot h_2 = 0}.
$$

Si resolvemos el sistema:

$$
v = \begin{pmatrix}x \ y \ z\end{pmatrix}, \quad
v \cdot h_1 = x + y = 0, \quad v \cdot h_2 = y + z = 0,
$$

obtenemos:

$$
y = -x, \quad z = -y = x \implies v = x \begin{pmatrix}1 \ -1 \ 1\end{pmatrix}.
$$

Por lo tanto, $H^\perp = \text{gen}{(1,-1,1)}$.

## Visualizacin en Plotly

Mostraremos:

* El plano $H$ en azul semitransparente.
* El vector $v = (2,0,1)$ en verde.
* La **proyeccin ortogonal de $v$ sobre $H$** en rojo.
* La componente en $H^\perp$ en morado.



````python
import numpy as np
import plotly.graph_objects as go

# =========================
# Vectores generadores del plano H
# =========================
h1 = np.array([1, 1, 0])
h2 = np.array([0, 1, 1])

# Vector a proyectar
v = np.array([2, 0, 1])

# =========================
# Base ortonormal para H usando Gram-Schmidt
# =========================
# Primer vector ortonormal
u1 = h1 / np.linalg.norm(h1)

# Segundo vector ortogonal a u1
proj_h2_u1 = np.dot(h2, u1) * u1
w2 = h2 - proj_h2_u1
u2 = w2 / np.linalg.norm(w2)

# =========================
# Proyeccin de v sobre H
# =========================
proj_v_H = np.dot(v, u1) * u1 + np.dot(v, u2) * u2

# Componente en H^perp
comp_v_Hperp = v - proj_v_H

# =========================
# Crear malla para el plano H
# =========================
s = np.linspace(-2, 2, 10)
t = np.linspace(-2, 2, 10)
S, T = np.meshgrid(s, t)
X = S * h1[0] + T * h2[0]
Y = S * h1[1] + T * h2[1]
Z = S * h1[2] + T * h2[2]

# =========================
# Crear figura
# =========================
fig = go.Figure()

# Plano H
fig.add_trace(go.Surface(x=X, y=Y, z=Z, colorscale=[[0, 'lightblue'], [1, 'lightblue']],
                         opacity=0.5, showscale=False, name='H'))

# Vector v
fig.add_trace(go.Scatter3d(x=[0,v[0]], y=[0,v[1]], z=[0,v[2]],
                           mode='lines+markers', line=dict(width=6, color='green'),
                           marker=dict(size=4), name='v'))

# Proyeccin sobre H
fig.add_trace(go.Scatter3d(x=[0, proj_v_H[0]], y=[0, proj_v_H[1]], z=[0, proj_v_H[2]],
                           mode='lines+markers', line=dict(width=6, color='red'),
                           marker=dict(size=4), name='proj_H(v)'))

# Componente en H^perp
fig.add_trace(go.Scatter3d(x=[proj_v_H[0], v[0]], y=[proj_v_H[1], v[1]], z=[proj_v_H[2], v[2]],
                           mode='lines+markers', line=dict(width=6, color='purple'),
                           marker=dict(size=4), name='componente en H^perp'))

# =========================
# Layout
# =========================
fig.update_layout(scene=dict(
                    xaxis_title='X',
                    yaxis_title='Y',
                    zaxis_title='Z',
                    aspectmode='cube'),
                  title="Complemento ortogonal de un subespacio H en $\mathbb{R}^3$"
                 )

fig.show()
````

## Teorema de aproximacin de la norma (Proyeccin ortogonal)

Sea $H$ un subespacio de $\mathbb{R}^{n}$ y $v \in \mathbb{R}^{n}$.

Sea $\operatorname{proj}_H(v)$ la **proyeccin ortogonal** de $v$ sobre $H$. Entonces:

[
|\ v - \operatorname{proj}*H(v)\ | = \min*{h \in H} |v - h|,
]

es decir, **$\operatorname{proj}_H(v)$ es el vector de $H$ ms cercano a $v$** en trminos de la norma euclidiana.

Equivalente geomtricamente:

[
v - \operatorname{proj}_H(v) \in H^\perp,
]

y para cualquier $h \in H$ diferente de $\operatorname{proj}_H(v)$, se cumple:

[
|v - h| > |v - \operatorname{proj}_H(v)|.
]

---

## Ejemplo en $\mathbb{R}^{2}$

Sea:

[
H = \text{la recta generada por } h = \begin{pmatrix}2 \ 1\end{pmatrix}, \quad
v = \begin{pmatrix}3 \ 2\end{pmatrix}.
]

La proyeccin ortogonal de $v$ sobre $H$ se calcula como:

[
\operatorname{proj}_H(v) = \frac{v \cdot h}{h \cdot h} , h
= \frac{3\cdot 2 + 2 \cdot 1}{2^2 + 1^2} \begin{pmatrix}2 \ 1\end{pmatrix}
= \frac{8}{5} \begin{pmatrix}2 \ 1\end{pmatrix}
= \begin{pmatrix}\frac{16}{5} \ \frac{8}{5}\end{pmatrix}.
]

Geomtricamente, $\operatorname{proj}_H(v)$ es el punto de la recta $H$ **ms cercano** a $v$.

## Desigualdad de Cauchy-Schwarz en (\mathbb{R}^{n})

Sea (u, v \in \mathbb{R}^{n}). Entonces se cumple:

[
|u \cdot v| \leq |u| , |v|,
]

donde (u \cdot v) es el producto escalar usual en (\mathbb{R}^{n}) y (|u|), (|v|) son las normas euclidianas de (u) y (v) respectivamente.

---

### Caso de igualdad

Se tiene igualdad si y solo si los vectores (u) y (v) son **linealmente dependientes**, es decir, si existe un escalar (\lambda \in \mathbb{R}) tal que

[
u = \lambda v \quad \text{o} \quad v = \lambda u.
]

---



Esta desigualdad es fundamental en geometra y lgebra lineal, y es la base para demostrar otras desigualdades como la **desigualdad triangular** y conceptos de **ngulo entre vectores** en (\mathbb{R}^{n}).
