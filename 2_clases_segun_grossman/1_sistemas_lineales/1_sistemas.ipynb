# Recorderis sobre la noción de linealidad  



Este libro trata del álgebra lineal. 

Al buscar la palabra “lineal” en el diccionario se encuentra, entre otras definiciones, la siguiente: lineal: (del lat. linealis). 1. adj. Perteneciente o relativo a la línea.

Sin embargo, en matemáticas la palabra “lineal” tiene un significado mucho más amplio. 

Una gran parte de la teoría de álgebra lineal elemental es, de hecho, una generalización de las propiedades de la línea recta. 

A manera de repaso se mencionan algunas propiedades fundamentales sobre las líneas rectas:

1. La pendiente $m$ de una recta que pasa por los puntos $(x_1, y_1)$ y $(x_2, y_2)$ está dada por

   $$
   m = \frac{y_2 - y_1}{x_2 - x_1}
   $$  




# Script en Python: pendiente de una recta que pasa por dos puntos




```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de los puntos
x1, y1 = 1, 2
x2, y2 = 4, 5

# Cálculo de la pendiente
m = (y2 - y1) / (x2 - x1)

print(f"Punto 1: ({x1}, {y1})")
print(f"Punto 2: ({x2}, {y2})")
print(f"Pendiente m = (y2 - y1) / (x2 - x1) = {m}")

# Definición de la recta que pasa por los dos puntos
x = np.linspace(min(x1, x2) - 1, max(x1, x2) + 1, 100)
y = m * (x - x1) + y1  # forma punto-pendiente

# Gráfica
plt.figure()
plt.plot(x, y, color = 'red', label = 'recta que pasa por P₁ y P₂')
plt.scatter([x1, x2], [y1, y2])
plt.axhline(0)
plt.axvline(0)
plt.grid(True)

# Etiquetas
plt.text(x1, y1, "  P₁(x₁, y₁)")
plt.text(x2, y2, "  P₂(x₂, y₂)")
plt.title("Pendiente de una recta que pasa por dos puntos")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.savefig(r"C:\Users\marco\Downloads\pendiente_recta.png")
plt.show()
```


## Interpretación didáctica (para los estudiantes)

* La **pendiente** mide el **cambio vertical** por cada **cambio horizontal**.
* En el código:

  * `y2 - y1` representa el **desplazamiento vertical**.
  * `x2 - x1` representa el **desplazamiento horizontal**.
* La ecuación usada para graficar la recta es la **forma punto–pendiente**:
  $$
  y - y_1 = m(x - x_1)
  $$

# 4. Exploración interactiva con Python



## 4.1 Importación de librerías



```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
from IPython.display import clear_output
```


# 4.2 Modelo interactivo: crecimiento del cultivo




```python
def pendiente_agro(x1, y1, x2, y2):
    
    # Limpiar salidas previas
    clear_output(wait=True)
    plt.close("all")
    
    if x1 == x2:
        print("La pendiente no está definida porque x₁ = x₂")
        return
    
    # Cálculo de la pendiente
    m = (y2 - y1) / (x2 - x1)
    
    print(f"Pendiente (tasa de crecimiento): {m:.2f} cm/día")
    
    # Recta
    x = np.linspace(min(x1, x2) - 1, max(x1, x2) + 1, 100)
    y = m * (x - x1) + y1
    
    # Gráfica
    fig, ax = plt.subplots()
    ax.axis([min(x)-1, max(x)+1, min(y)-1, max(y)+1])
    ax.plot(x, y)
    ax.scatter([x1, x2], [y1, y2])
    ax.grid(True)
    
    ax.set_title("Crecimiento del cultivo (modelo lineal)")
    ax.set_xlabel("Tiempo (días)")
    ax.set_ylabel("Altura promedio (cm)")
    
    ax.text(x1, y1, f"  ({x1}, {y1})")
    ax.text(x2, y2, f"  ({x2}, {y2})")
    
    plt.show()
```



# 4.3 Sliders interactivos (experimentación)




```python
interact(
    pendiente_agro,
    x1=widgets.IntSlider(value=2, min=0, max=30, step=1, description='Día x₁'),
    y1=widgets.IntSlider(value=20, min=0, max=200, step=5, description='Altura y₁'),
    x2=widgets.IntSlider(value=10, min=1, max=40, step=1, description='Día x₂'),
    y2=widgets.IntSlider(value=80, min=0, max=300, step=5, description='Altura y₂')
);
```

# * Si $x_2 - x_1 = 0$ y $y_2 \neq y_1$, entonces la recta es vertical y se dice que la pendiente es indefinida.

Aquí tienes un **script claro y didáctico en Python**, pensado para **Jupyter Notebook**, que **grafica una recta vertical** y **explicita por qué la pendiente es indefinida** cuando
$$
x_2 - x_1 = 0 \quad \text{y} \quad y_2 \neq y_1.
$$

Este ejemplo encaja muy bien como **caso límite** dentro de tu secuencia ABP.

# Script en Python: recta vertical (pendiente indefinida)



```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de los puntos con la misma coordenada x
x1, y1 = 3, 2
x2, y2 = 3, 8   # x2 - x1 = 0 y y2 ≠ y1

print("Caso: recta vertical")
print(f"Punto 1: ({x1}, {y1})")
print(f"Punto 2: ({x2}, {y2})")
print("La pendiente es indefinida porque x2 - x1 = 0")

# Valores de y para trazar la recta vertical
y = np.linspace(min(y1, y2) - 2, max(y1, y2) + 2, 100)
x = np.full_like(y, x1)

# Gráfica
plt.figure()
plt.plot(x, y, color='red', label='recta vertical x = 3')
plt.scatter([x1, x2], [y1, y2])
plt.axhline(0)
plt.axvline(0)
plt.grid(True)

plt.title("Recta vertical: pendiente indefinida")
plt.xlabel("x")
plt.ylabel("y")

plt.text(x1, y1, f"  P₁({x1},{y1})")
plt.text(x2, y2, f"  P₂({x2},{y2})")

plt.show()
```


# Interpretación matemática

* La fórmula
  $$
  m = \frac{y_2 - y_1}{x_2 - x_1}
  $$
  **no se puede aplicar** porque el denominador es cero.
* Geométricamente:

  * no hay “avance horizontal”,
  * todo el cambio es vertical,
  * la recta **no puede escribirse** como ( y = mx + b ).

---

* Cualquier recta (a excepción de aquella que tiene una pendiente indefinida) se puede 
describir con su ecuación en la forma pendiente-ordenada al origen $y = mx + b$, donde 
$m$ es la pendiente de la recta y $b$ es la ordenada al origen (el valor de $y$ en el punto en el que la recta cruza el eje $y$).

A continuación te dejo un **script completo y estable en Python**, usando **matplotlib + ipywidgets**, para **visualizar interactivamente la forma pendiente–ordenada al origen**

$$
y = mx + b
$$

excluyendo explícitamente el caso de **pendiente indefinida**. El diseño es **didáctico**, pensado para **precálculo / cálculo** y coherente con tu secuencia ABP.

## Celda inicial (ejecutar una sola vez)


```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
from IPython.display import clear_output
```

# Función interactiva: forma pendiente–ordenada


```python
def recta_pendiente_ordenada(m, b):
    
    clear_output(wait=True)
    plt.close("all")
    
    # Dominio
    x = np.linspace(-10, 10, 200)
    y = m * x + b
    
    # Gráfica
    fig, ax = plt.subplots()
    ax.plot(x, y, color='red')
    ax.axis([-10, 10, -10, 10])
    ax.axhline(0)
    ax.axvline(0)
    ax.grid(True)
    
    # Punto de corte con el eje y
    ax.scatter(0, b)
    ax.text(0, b, f"  (0, {b})")
    
    ax.set_title("Recta en forma pendiente–ordenada:  y = mx + b")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    
    # Información didáctica
    ax.text(
        -9, 9,
        f"m = {m}\nb = {b}",
        bbox=dict(boxstyle="round")
    )
    
    plt.show()
```


# Sliders interactivos


```python
interact(
    recta_pendiente_ordenada,
    m=widgets.FloatSlider(
        value=1,
        min=-5,
        max=5,
        step=0.5,
        description='Pendiente m'
    ),
    b=widgets.FloatSlider(
        value=0,
        min=-10,
        max=10,
        step=1,
        description='Ordenada b'
    )
);
```


## Interpretación matemática (para guiar la clase)

* El **parámetro $m$** controla:

  * inclinación de la recta,
  * crecimiento ($ m > 0 $) o decrecimiento ($ m < 0 $).
* El **parámetro $ b $** determina:

  * el punto donde la recta corta el eje $ y $,
  * traslaciones verticales.
* Este modelo **no incluye** rectas verticales, porque:
  $$
  x = c \quad \text{no puede escribirse como } y = mx + b
  $$

# * Dos rectas distintas son paralelas si y sólo si tienen la misma pendiente.

Aquí tienes una **ilustración interactiva con `matplotlib` e `ipywidgets`** que muestra claramente la afirmación:

> **Dos rectas distintas son paralelas si y solo si tienen la misma pendiente.**

La idea es:

* fijar **la misma pendiente (m)** para ambas rectas,
* permitir **ordenadas al origen distintas** $b_1 \neq b_2$,
* observar que **no se cortan** ⇒ son paralelas.

---



# Celda inicial (ejecutar una sola vez)



```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
from IPython.display import clear_output
```


# Función interactiva: dos rectas paralelas


```python
def rectas_paralelas(m, b1, b2):
    
    clear_output(wait=True)
    plt.close("all")
    
    x = np.linspace(-10, 10, 200)
    
    y1 = m * x + b1
    y2 = m * x + b2
    
    fig, ax = plt.subplots()
    ax.plot(x, y1, label=f"y = {m}x + {b1}", color ='red')
    ax.plot(x, y2, label=f"y = {m}x + {b2}", color ='green')
    
    ax.axhline(0)
    ax.axvline(0)
    ax.grid(True)
    
    ax.set_title("Dos rectas paralelas tienen la misma pendiente")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.axis([-10, 10, -10, 10])
    ax.legend()
    
    # Mensaje didáctico
    if b1 != b2:
        texto = "Las rectas son distintas y paralelas\n(misma pendiente, distinto intercepto)"
    else:
        texto = "Las rectas coinciden\n(misma pendiente y mismo intercepto)"
    
    ax.text(
        -9, 9,
        texto,
        bbox=dict(boxstyle="round")
    )
    
    plt.show()
```



# Sliders interactivos


```python
interact(
    rectas_paralelas,
    m=widgets.FloatSlider(
        value=1,
        min=-5,
        max=5,
        step=0.5,
        description='Pendiente m'
    ),
    b1=widgets.FloatSlider(
        value=2,
        min=-10,
        max=10,
        step=1,
        description='b₁'
    ),
    b2=widgets.FloatSlider(
        value=-3,
        min=-10,
        max=10,
        step=1,
        description='b₂'
    )
);
```


## Lectura matemática guiada

* Si dos rectas tienen ecuaciones:
  $$
  y = m x + b_1 \quad \text{y} \quad y = m x + b_2
  $$
  entonces:

  * **misma pendiente** $m$,
  * **ordenadas distintas** $b_1 \neq b_2$,
  * ⇒ **rectas paralelas**.

* Si además $b_1 = b_2$:

  * no son dos rectas distintas,
  * es **la misma recta**.

# * Si la ecuación de la recta se escribe en la forma $ax + by = c$, ($b \neq 0$), entonces se puede calcular fácilmente la pendiente $m$, como $m = -a/b$. 

# * Si $m_1$ es la pendiente de la recta $L_1$, $m_2$ es la pendiente de la recta $L_2$, $m_1 \neq 0$ y $L_1$ y $L_2$ son perpendiculares, entonces $m_2 = -1/m1$.

A continuación te dejo un **script completo, estable y didáctico en Python**, usando **`matplotlib` + `ipywidgets`**, para **ilustrar interactivamente la condición de perpendicularidad entre dos rectas**:

$$
\text{Si } L_1 \perp L_2,; m_1 \neq 0 \quad \Rightarrow \quad m_2 = -\frac{1}{m_1}
$$

El diseño evita el caso problemático $m_1 = 0$ y hace explícita la relación entre pendientes.

---



## Celda inicial (ejecutar una sola vez)


```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
from IPython.display import clear_output
```


# Función interactiva: rectas perpendiculares


```python
def rectas_perpendiculares(m1, b1, b2):
    
    clear_output(wait=True)
    plt.close("all")
    
    # Evitar el caso m1 = 0
    if m1 == 0:
        print("⚠️ m₁ no puede ser 0 para definir una recta perpendicular.")
        return
    
    # Pendiente perpendicular
    m2 = -1 / m1
    
    # Dominio
    x = np.linspace(-10, 10, 200)
    
    y1 = m1 * x + b1
    y2 = m2 * x + b2
    
    # Gráfica
    fig, ax = plt.subplots()
    ax.plot(x, y1, label=f"L₁: y = {m1}x + {b1}", color ='red')
    ax.plot(x, y2, label=f"L₂: y = ({m2:.2f})x + {b2}", color ='green')
    
    ax.axhline(0)
    ax.axvline(0)
    ax.grid(True)
    
    ax.set_aspect("equal", adjustable="box")
    
    ax.set_title("Rectas perpendiculares")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    
    ax.legend()
    
    # Información didáctica
    ax.text(
        -9, 9,
        f"m₁ = {m1}\nm₂ = -1/m₁ = {m2:.2f}",
        bbox=dict(boxstyle="round")
    )
    ax.axis([-10, 10, -10, 10])
    plt.show()
```


# Sliders interactivos



```python
interact(
    rectas_perpendiculares,
    m1=widgets.FloatSlider(
        value=1,
        min=-5,
        max=5,
        step=0.5,
        description='Pendiente m₁'
    ),
    b1=widgets.FloatSlider(
        value=0,
        min=-10,
        max=10,
        step=1,
        description='Intercepto b₁'
    ),
    b2=widgets.FloatSlider(
        value=2,
        min=-10,
        max=10,
        step=1,
        description='Intercepto b₂'
    )
);
```

# Lectura matemática guiada

* Dos rectas:
  $$
  L_1: y = m_1 x + b_1,\qquad
  L_2: y = m_2 x + b_2
  $$
* Son **perpendiculares** si:
  $$
  m_1 \cdot m_2 = -1
  $$
* El slider muestra que:

  * al variar $m_1$,
  * $m_2$ se ajusta automáticamente como $-1/m_1$,
  * el ángulo entre las rectas es de $90^\circ$.

## TAREA 5. Paralelismo y perpendicularidad

**Objetivo:** Fijar relaciones entre pendientes.

### Instrucciones

1. Dadas las rectas:

   * $ L_1: y = 3x + 1 $
   * $ L_2: y = 3x - 4 $
   * $ L_3: y = -\frac{1}{3}x + 2 $

2. Determina:

   * cuáles son paralelas,
   * cuáles son perpendiculares,
   * justifica usando las pendientes.
